```r
college$Private <- as.factor(college$Private)
```

```r
plot(college$Private, college$Outstate,
     xlab = "Outstate",
     ylab = "Private (Yes/No)",
     horizontal = TRUE)
```

```r
Elite = rep('No', nrow(college))
Elite[college$Top10perc > 50] = 'Yes'
Elite = as.factor(Elite)
college = data.frame(college, Elite)
```

```r
summary(college)
```

```r
plot(college$Elite, college$Outstate,
     xlab = "Outstate",
     ylab = "Elite (Yes/No)",
     horizontal = TRUE)
```

```r
par(mfrow = c(2, 2))

# Histogram for Outstate with 10 bins
hist(college$Outstate,
     main = "Histogram of Outstate Tuition",
     xlab = "Outstate Tuition (USD)",
     col = "lightblue",
     breaks = 10)

# Histogram for Apps with 20 bins
hist(college$Apps,
     main = "Histogram of Applications",
     xlab = "Number of Applications",
     col = "lightgreen",
     breaks = 20)

# Histogram for Room.Board with 15 bins
hist(college$Room.Board,
     main = "Histogram of Room & Board Costs",
     xlab = "Room & Board (USD)",
     col = "lightcoral",
     breaks = 15)

# Histogram for Grad.Rate with 5 bins
hist(college$Grad.Rate,
     main = "Histogram of Graduation Rate",
     xlab = "Graduation Rate (%)",
     col = "lightgoldenrod",
     breaks = 5)
```
