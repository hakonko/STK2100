```r
# a)
attach(Auto)
model <- lm(mpg ~ horsepower)
summary(model)
```

**Is there a relationship between the predictor and response?**\
* Yes, there is. A low p-score of `<2e-16` for the `horsepower` coefficient, strongly suggests that `horsepower` has statistically significant effect on `mpg`.
* The t-value is large in magnitude (`-24.49`), indicating a strong relationship.
* The R-squared value of `0.6059` means that about 60% of the variance in mpg can be described by `horsepower` alone
* The negative coefficient (`-0.1578`) suggests that as `horsepower` increases, `mpg` decreases, meaning that cars with more hp tends to be less fuel efficient

```r
new_data <- data.frame(horsepower = 98)
confint_mpg <- predict(model, new_data, interval="confidence", level=0.98)
predint_mpg <- predict(model, new_data, interval="prediction")
print(confint_mpg)
print(predint_mpg)
```

**Confidence interval:**\
Good for estimating the average mpg for all cars with horsepower = 98. Gives an estimat on how certain the model is about the mean prediction. The CI is usually narrower because it only accounts for the uncertainty in estimating the mean.\
**Prediction interval:**\
Good for predicting the mpg of a new, unseen car with horsepower = 98. PI is wider than CI because it accounts for both model uncertainty and natural variation among individual cars. This is crucial when making individual predictions.
